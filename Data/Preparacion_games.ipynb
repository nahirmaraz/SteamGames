Teniendo en cuenta las funciones que hay que hacer para consumo de la API, se consideran irrelevantes las columnas 'publisher', 'title', 'url', 'tags', 'reviews_url', 'specs', 'price' y 'early_acces'.

```python
df_g=df_g.drop(['publisher','title','url', 'tags','reviews_url','specs','price','early_access'], axis=1)
```

Considerando que los valores nulos pueden llevar a un error y mala interpretacion de los datos en las columnas 'app_name','id', 'release_date'; y tambien la cantidad de nulos es peque√±a, los eliminamos.

```python
df_g=df_g.dropna(subset=['app_name','id', 'release_date'])
df_g.head(4)
```

```python
df_g['genres'].fillna('No se conoce', inplace=True)
df_g['developer'].fillna('No se conoce', inplace=True)
df_g.info()
```

```python
# Reemplazar el valor 0 en la columna 'genres' con 'No se conoce'
df_g['genres'] = df_g['genres'].replace(0, 'No se conoce')
```

```python
df_g['release_date'] = pd.to_datetime(df_g['release_date'], format='%Y-%m-%d', errors='coerce')
print('N√∫mero de registros en formato incorrecto:', df_g['release_date'].isna().sum())
```
